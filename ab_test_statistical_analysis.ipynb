```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from statsmodels.stats.proportion import proportions_ztest
```

```python
df = pd.read_csv('/Users/skorykkateryna/Downloads/ab_test_results.csv')
df.head () 
```

```python
# 2. Split data into control and treatment groups
control = df[df['variant'] == 'control']
treatment = df[df['variant'] == 'treatment']

# Extract sample sizes
n_control = len(control)
n_treatment = len(treatment)

print("=== A/B TEST STATISTICAL RESULTS ===\n")
```

```python
# --- TEST 1: Z-test for Conversions ---
# Count successful conversions for both groups
conv_control = control['converted'].sum()
conv_treatment = treatment['converted'].sum()

# Perform two-sample proportions Z-test
z_stat, p_val_conv = proportions_ztest(
    count=[conv_treatment, conv_control], 
    nobs=[n_treatment, n_control]
)

print(f"1. Conversion Rate (Z-test):")
print(f"   Control CR: {control['converted'].mean():.2%}")
print(f"   Treatment CR: {treatment['converted'].mean():.2%}")
print(f"   Z-score = {z_stat:.5f}")
print(f"   p-value = {p_val_conv}")  # Removed :.5f because if p-value is extremely small (like 1e-8), :.5f will just show 0.00000
print(f"   Statistical Significance (alpha=0.05): {p_val_conv < 0.05}\n")
```

```python
# --- TEST 2: T-test for Session Duration ---
# Perform Welch's T-test (equal_var=False handles potential variance differences)
t_stat_time, p_val_time = stats.ttest_ind(
    treatment['session_duration'], 
    control['session_duration'], 
    equal_var=False
)

print(f"2. Session Duration (Welch's T-test):")
print(f"   Control Mean: {control['session_duration'].mean():.3f} min")
print(f"   Treatment Mean: {treatment['session_duration'].mean():.3f} min")
print(f"   t-score = {t_stat_time:.5f}")
print(f"   p-value = {p_val_time}")
print(f"   Statistical Significance (alpha=0.05): {p_val_time < 0.05}\n")
```

```python
# --- TEST 3: T-test for Pages Visited ---
# Perform Welch's T-test for discrete/continuous metric of engagement
t_stat_pages, p_val_pages = stats.ttest_ind(
    treatment['pages_visited'], 
    control['pages_visited'], 
    equal_var=False
)

print(f"3. Pages Visited (Welch's T-test):")
print(f"   Control Mean: {control['pages_visited'].mean():.3f}")
print(f"   Treatment Mean: {treatment['pages_visited'].mean():.3f}")
print(f"   t-score = {t_stat_pages:.5f}")
print(f"   p-value = {p_val_pages}")
print(f"   Statistical Significance (alpha=0.05): {p_val_pages < 0.05}")
```
